# Top 250 Nifty Stocks - Web Scrapping

```python
URL = "https://www.equitymaster.com/stockquotes/1-76/list-of-nifty-largemidcap-250-stocks"
```

```python
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:127.0) Gecko/20100101 Firefox/127.0',
    'Accept-Language': 'en-US,en;q=0.5'
}
```

```python
import requests
try:
    webpage = requests.get(URL,headers=HEADERS)
    print(webpage)  # This will print the response object
except Exception as err:
    print(f"Other error occurred: {err}")
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(bytes(webpage.content),'html.parser')
soup
```

```python
table = soup.find("table",attrs = {"class":"mystocks cls"})
```

```python
table_head = soup.find("table",attrs = {"class":"mystocks cls"}).thead.tr.find_all("th")
table_head
```

```python
# Extract and print the text from each header
headings = []
for head in table_head:
    headings.append(head.get_text(strip=True))  # Use strip=True to remove extra whitespace

print(headings)
```

```python
table_row = soup.find("table",attrs = {"class":"mystocks cls"}).tbody.find_all("tr")
table_row
```

```python
scrip = []
bse_price = []
bse_percentage = []
nse_price = []
nse_percentage = []


for records in table_row:
    
    scrip.append(records.td.a.get_text(strip=True))
    bse_price.append(records.contents[3].contents[0].text.strip())  # Contents refers to the html DOM tree childrens
    bse_percentage.append(records.contents[3].contents[1].text.strip()) 
    nse_price.append(records.contents[5].contents[0].text.strip()) 
    nse_percentage.append(records.contents[5].contents[1].text.strip()) 
```

```python
nse_percentage
```

```python
# Create a DataFrame
import pandas as pd
df = pd.DataFrame({
    'scrip': scrip,
    'bse_price': bse_price,
    'bse_percentage': bse_percentage,
    'nse_price': nse_price,
    'nse_percentage': nse_percentage
}) 
df
```

### Advanced web scrapping using selenium

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# Set up Selenium WebDriver to use system installed Chrome
options = Options()
options.headless = False  # Set to True if you want to run in headless mode
options.add_argument("--start-maximized")  # Maximize the Chrome window on open (optional)
options.add_argument("--disable-extensions")  # Disable extensions (optional)
options.add_argument("--disable-gpu")  # Disable GPU acceleration (optional)
options.add_argument("--no-sandbox")  # Bypass OS security model (optional)

# Initialize Chrome WebDriver with system path
driver = webdriver.Chrome(options=options)

# URL of the page to scrape

driver.get(URL)

# Wait for the page to fully load (adjust as needed)
time.sleep(5)

# Parse the fully rendered page content using BeautifulSoup (if needed)
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all td tags with the specified class
td_elements = soup.find_all('td', class_='msloss alignright')

# Extract and store the text separately
main_texts = []
small_texts = []
for td in td_elements:
    main_text = td.contents[0].strip()  # Extract the main text (excluding the small tag)
    small_text = td.find('small').get_text(strip=True)  # Extract the text within the small tag
    main_texts.append(main_text)
    small_texts.append(small_text)

# Print the results
print("Main texts:", len(main_texts))
print("Small texts:", small_texts)

# Close the WebDriver
driver.quit()
```

### There are several reasons for not get all records or encounter limitations:

1. **Pagination**: If the data you're scraping is spread across multiple pages, you need to handle pagination correctly to scrape all pages.

2. **JavaScript Rendering**: Some websites use JavaScript to load content dynamically. Libraries like `requests` in Python may not handle this by default, requiring tools like `Selenium` for more complex scraping.

3. **Anti-Scraping Measures**: Websites may employ techniques like CAPTCHAs, rate limiting, or IP blocking to prevent automated scraping.

4. **Session Handling**: If the website requires login or session cookies to access certain content, you need to manage sessions appropriately in your scraping script.

5. **Structure Changes**: Websites frequently update their HTML structure, CSS classes, or API endpoints, breaking existing scrapers. Regular maintenance and updates to your scraper may be necessary.

6. **Data Volume**: Scraping large volumes of data can be time-consuming and resource-intensive. Sometimes, incomplete data retrieval may be due to timeouts or network issues.

To overcome these limitations, ensure your scraping script is robust, handles errors gracefully, respects website policies (like robots.txt), and implements retries and delays where appropriate.
